```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('dataset - 2020-09-24.csv')
```

```python
dataset
```

```python
df = dataset.copy()
```

```python
df
```

```python
for col in df.columns:
    print(col)
```

# Proses Preprocessing Data

```python
df.fillna(value = 0)
```

```python
df_new = df.loc[:571,["Name", "Age", "Goals"]]
```

```python
df_new = df_new.fillna(value = 0.0)
df_new
```

```python
dataset = []

for i in range(0,571):
    dataset.append([float(df_new.values[i,j]) for j in range(1,3)])
    
```

```python
dataset
```

```python
arrayDataset = np.array(dataset)
arrayDataset
```

```python
points = arrayDataset
```

# Implementasi kedalam Algortima KMeans

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans.fit(dataset)
```

*Data sebelum clustering*

```python
plt.scatter(arrayDataset[:,0], arrayDataset[:,1])
```

*Penentuan Center dari tiap tiap Cluster*

```python
clusters = kmeans.cluster_centers_
clusters
```

*Melihat posisi cluster dari tiap tia point*

```python
y_km = kmeans.fit_predict(points)
```

```python
y_km
```

*Hasil*

```python
plt.scatter(points[y_km == 0,0], points[y_km == 0,1], s=50, color = 'red')
plt.scatter(points[y_km == 1,0], points[y_km == 1,1], s=50, color = 'green')
plt.scatter(points[y_km == 2,0], points[y_km == 2,1], s=50, color = 'yellow')

plt.scatter(clusters[0][0], clusters[0][1], marker='*', s=200, color = 'black')
plt.scatter(clusters[1][0], clusters[1][1], marker='*', s=200, color = 'black')
plt.scatter(clusters[2][0], clusters[2][1], marker='*', s=200, color = 'black')

plt.show()
```
